# Preprocessing the csv file:

## Importing libraries:


```python
import pandas as pd
import numpy as np
import matplotlib as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow import keras
from tensorflow.keras.callbacks import EarlyStopping
```

## Reading the data

```python
data = pd.read_csv('train.csv')
data
```

## Make the data ready for the model:
### 1) Drop (user_uuid , restaurant_features) columns -*Useless*- :

```python
data = data.drop('user_uuid' , axis=1)
data = data.drop('restaurant_uuid' , axis=1)
data
```

### 2) cuisine and restaurant_features columns are strings not list:

```python
data['cuisine'] , data['restaurant_features']
```

### Make function convert these columns:

```python
def StrToList(s):
    s = s.apply(lambda x : x[1:-1].replace("' '", ","))
    s = s.apply(lambda x : x.replace("'",""))
    s = s.apply(lambda x : x.lower())
    s = s.apply(lambda x: x.split(","))
    return s
```

### Update columns:

```python
data['cuisine']=StrToList(data['cuisine'])
data['restaurant_features']=StrToList(data['restaurant_features'])
```

### Make function check for errors in column:

```python
def catch_err(s):
    mr = []
    for i in range(len(s)):
        for j in range(len(s[i])):
            if "[" in s[i][j]:
                mr.append(i)
    mr=pd.Series(mr)
    return mr
```

### Get unique errors

```python
catch_err(data['restaurant_features'])
```

```python
errors = pd.concat([catch_err(data['cuisine']),catch_err(data['restaurant_features'])]).unique()
print("Num of errors=",len(errors))
print("Num of data=",len(data))
print("% of error=",(len(errors)/len(data))/100)
```

#### Small num or errors we can just drop them

```python
data = data.drop(errors)
```

### Reset the index of Data:

```python
data.reset_index(drop=True , inplace=True)
```

### Check the unique values

```python
cu = pd.Series([x for l in data['cuisine'] for x in l]).unique()
cu
```

```python
ru = pd.Series([x for l in data['restaurant_features'] for x in l]).unique()
ru
```

### Clear the /r/n in restaurant_features column:

```python
restaurant_temp = data['restaurant_features']
```

```python
restaurant_temp = restaurant_temp.apply(lambda x : str(x))
restaurant_temp = restaurant_temp.apply(lambda x : x.replace("\\r\\n" , ","))
restaurant_temp = restaurant_temp.apply(lambda x : x.replace("'" , ""))
restaurant_temp = restaurant_temp.apply(lambda x : x.replace(" ",""))
data['restaurant_features'] = restaurant_temp.apply(lambda x : x[1:-1].split(","))
```

### Recheck unique values in restaurant_features:

```python
ru = pd.Series([x for l in data['restaurant_features'] for x in l]).unique()
ru
```

### Convert all columns to dummies data frames:

```python
area_dumm = pd.get_dummies(data['area'])
cuisine_dumm = pd.get_dummies(data['cuisine'].apply(pd.Series).stack()).groupby(level=0).sum()
rest_f_dumm = pd.get_dummies(data['restaurant_features'].apply(pd.Series).stack()).groupby(level=0).sum()
```

### Recombine all dummies data frames in 1 Data frame:

```python
f_data = pd.concat([data['discounted'] , area_dumm , data['nOrders'] , cuisine_dumm , rest_f_dumm , data['Ratings']], axis=1)
f_data
```

# THIS IS THE FINAL DATA FRAME:

```python
f_data
```

# Define X,Y and split data

```python
#X,Y
X = f_data.drop('Ratings' , axis=1)
Y = np.zeros((len(data),5))
for i in range(len(data)):
    Y[i, f_data['Ratings'][i]-1]=1
#Train , Test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 215)
```

### X:

```python
X
```

```python
Y
```

### Train and test lengths:

```python
print("length Train" , len(X_train))
print("length Test" , len(X_test))
```

# Make the structure of the model:

```python
model=keras.models.Sequential([
    keras.layers.InputLayer(input_shape=(118)),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(5, activation='softmax')

])
```

# Optimizer , compile and fitting of the model:

```python
optimizer=tf.keras.optimizers.Adam(0.001)
es = EarlyStopping(monitor='accuracy' , mode='max' , patience=5)
model.compile(optimizer='adam' , loss=tf.keras.losses.categorical_crossentropy ,metrics=['accuracy'])
model.fit(X_train,Y_train, epochs=100 , callbacks=[es]) 
```

## Check the Test data:

```python
model.evaluate(X_test,Y_test)
```
